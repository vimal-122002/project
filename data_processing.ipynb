# New Section

```python
 #databinning
df['horsepower']=df['horsepower'].astype(float)
df['horsepower'].dtypes
```

```python
#visualising horsersepower data in histogram
```

```python
import matplotlib.pyplot as plt
plt.hist(df['horsepower'])
```

```python
#create bins
```

```python
bins = np.linspace(min(df['horsepower']) , max(df['horsepower']), 4)
bins
```

```python
groups_name = ['Low' , 'Medium' , 'High']
```

```python
df['horsepower_binned'] = pd.cut(df['horsepower'] , bins , labels = groups_name , include_lowest = True)
df[['horsepower' , 'horsepower_binned']].head(10)
```

```python
p = df['horsepower_binned'].value_counts()
p
```

```python
plt.bar(groups_name , p)
plt.xlabel("HorsePower")
plt.ylabel("Count")
```
